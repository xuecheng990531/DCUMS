```python
import numpy as np
from glob import glob
import pickle
```

```python
ALL_IMAGES_DIR = sorted([f for f in glob(r"/icislab/volume1/lxc/DCUMS/data/cvc/images/*") if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff'))])
ALL_MASK_DIR = sorted([f for f in glob(r"/icislab/volume1/lxc/DCUMS/data/cvc/masks/*") if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff'))])
```

```python
# 20/20/60 train test unlabelled
# np.random.shuffle(ALL_IMAGES_DIR)
n = len(ALL_IMAGES_DIR)
print("Total number of images:", n)
n_train = int(0.2 * n)
n_test  = int(0.2 * n)
# 剩下的都给未标注
train_val_img   = ALL_IMAGES_DIR[:n_train]
test_img        = ALL_IMAGES_DIR[n_train:n_train+n_test]
unlabelled_img  = ALL_IMAGES_DIR[n_train+n_test:]

train_val_label = []

for i in train_val_img:
    train_val_label.append(i.replace("images", "masks"))
test_label = []
for i in test_img:
    test_label.append(i.replace("images", "masks"))
unlabelled_label = []
for i in unlabelled_img:
    unlabelled_label.append(i.replace("images", "masks"))

# files contain paths of slices chosen for each set. All pre-processed samples can be found in the 'all_data' folder
pickle.dump(train_val_img, open('/icislab/volume1/lxc/DCUMS/data/cvc/train_val/img/train_val.data', 'wb'), -1)
pickle.dump(train_val_label, open('/icislab/volume1/lxc/DCUMS/data/cvc/train_val/label/train_val.mask', 'wb'), -1)
pickle.dump(test_img, open('/icislab/volume1/lxc/DCUMS/data/cvc/test/img/test.data', 'wb'), -1)
pickle.dump(test_label, open('/icislab/volume1/lxc/DCUMS/data/cvc/test/label/test.mask', 'wb'), -1)
pickle.dump(unlabelled_img, open('/icislab/volume1/lxc/DCUMS/data/cvc/unlabelled/img/unlabelled.data', 'wb'), -1)
pickle.dump(unlabelled_label, open('/icislab/volume1/lxc/DCUMS/data/cvc/unlabelled/label/unlabelled.mask', 'wb'), -1)
```
